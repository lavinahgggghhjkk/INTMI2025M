```python
horas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas <= 40:
    salario = horas * valor_hora
else:
    extras = horas - 40
    salario = (40 * valor_hora) + (extras * valor_hora * 1.5)

print(f"Salário total: R$ {salario:.2f}")
```

## Jogo da adivinhação

```python
import random as rd
```

```python
aleatorio = rd.randint(0,100)

print("===========================================")
print("            JOGO DE ADIVINHAÇÃO            ")
print("===========================================")
print("Tente adivinhar o número que estou pensando")
print("===========================================")

tentativa = 10

while (tentativa > 0):
  numero = int(input("Digite um númeo: "))
  if (numero == aleatorio):
    print("Parabés! Você acertou!")
    break
  elif (numero > aleatorio):
    print("Digite um número menor:")
  else:
    print("Digite um número maior:")
  tentativa -= 1
if (tentativa == 0):
  print("Você não adivinhou!")
```
